```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:/Users/USER/Downloads/archive (1)/site_hourly_usage_Jakande1.csv", skiprows=4)
df
```

```python
df['Dates'] = pd.to_datetime(df['Timestamp (UTC)']).dt.date
df['Time'] = pd.to_datetime(df['Timestamp (UTC)']).dt.time
```

```python
df
```

```python
df['Dates'] = df['Dates'].astype(str)
df=df[df['Dates']=='2021-03-17']
df
```

```python
df_usage = pd.pivot_table(df,index=['Time'],values=['Usage Amount'], aggfunc=np.sum)
```

```python
df_usage
```

```python
df_usage['Time'] = df_usage['Time'].astype(str)


plt.figure(figsize=(15,10))
plt.plot(df_usage['Time'], df_usage["Usage Amount"])
plt.xticks(rotation=90)

plt.show()
```
